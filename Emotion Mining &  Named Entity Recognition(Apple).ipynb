# Emotion Mining

```python
import numpy as np
import pandas as pd
import string
import spacy
from matplotlib.pyplot import imread
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
afinn=pd.read_csv("Downloads\\Afinn.csv",sep=",",encoding='latin-1')
afinn
```

```python
afinn.shape
```

```python
afinn.head(20)
```

```python
book=pd.read_csv("Downloads\\apple.txt",error_bad_lines=False)
book
```

```python
book=[x.strip() for x in book.x]
book=[x for x in book if x]
book[0:10]
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk import tokenize
sentences=tokenize.sent_tokenize("".join(book))
sentences[0:10]
```

```python
sent_df=pd.DataFrame(sentences,columns=['sentence'])
sent_df
```

```python
affinity_scores=afinn.set_index('word')['value'].to_dict()
affinity_scores
```

```python
nlp=spacy.load('en_core_web_md')
sentiment_lexicon=affinity_scores

def calculate_sentiment(text:str=None):
    sent_score=0
    
    if text:
        sentence=nlp(text)
        for word in sentence:
            sent_score+=sentiment_lexicon.get(word.lemma_,0)
    return sent_score    
```

```python
calculate_sentiment(text='I love coding,but it is very difficult sometimes,though it is intresting too')
```

```python
sent_df['sentiment_value']=sent_df['sentence'].apply(calculate_sentiment)
```

```python
sent_df['word_count']=sent_df['sentence'].str.split().apply(len)
sent_df.head(10)
```
